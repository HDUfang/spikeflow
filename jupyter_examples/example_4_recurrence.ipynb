# recurrent layer

```python
import numpy as np
from spikeflow import BPNNModel, IzhikevichNeuronLayer
from spikeflow import Synapse, ComplexSynapseLayer, weights_from_synapses
import spikeflow.drawing_utils.trace_renderers as rend
```

# Create a model layer with 3 neurons in a single layer, with 1 recurrent connection

```python
model_input_shape = (3,)

neuron_layer = IzhikevichNeuronLayer.layer_from_tuples('n1', [
    IzhikevichNeuronLayer.C(a=0.010, b=0.2, c=-65.0, d=6.0, t=30.0, v0=0.0),
    IzhikevichNeuronLayer.C(a=0.015, b=0.2, c=-65.0, d=6.0, t=30.0, v0=0.0),
    IzhikevichNeuronLayer.C(a=0.020, b=0.2, c=-65.0, d=6.0, t=30.0, v0=0.0)
])

# A single decaying synapse, from neuron 0 to neuron 2 in same layer
recurrent_synapse_layer = ComplexSynapseLayer('rsyn',
    neuron_layer,
    neuron_layer,
    decay = 0.95, 
    weights = weights_from_synapses(
        neuron_layer, 
        neuron_layer,
        [ Synapse(from_neuron_index=0, to_neuron_index=2, v=7.0) ]
    )
)

model = BPNNModel.compiled_model(model_input_shape, [neuron_layer], [recurrent_synapse_layer])
```

# Run the model for 2000 timesteps

```python
traces = []

def end_time_step_callback(i, graph, sess, results):
    traces.append(results)
    
data = (np.ones(3,)*(7 if i > 1200 else 0) for i in range(0, 2000))
    
model.run_time(data, end_time_step_callback)
```

# Extract the data we want and display (after 1000 timestep settle-down period)

```python
neuron_layer_0_traces = np.array([r['n1'] for r in traces])

rend.render_figure([rend.IzhikevichNeuronTraceRenderer(neuron_layer_0_traces, 'Layer 0 Neuron')], 1000, 2000)
```

In final graph above, we can see that neuron 2 fires immediately after the first firing, due to the recurrent synapse
