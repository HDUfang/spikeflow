# Three Leaky Integrate-and-fire neurons

```python
import numpy as np
from spikeflow import BPNNModel, LIFNeuronLayer
import spikeflow.drawing_utils.trace_renderers as rend
```

# Create a model layer with 3 neurons in a single layer

```python
model_input_shape = (3,)

nl = LIFNeuronLayer.layer_from_tuples([
    LIFNeuronLayer.C(r=1.0, c=10.0, t=1.0, refrac=30),
    LIFNeuronLayer.C(r=1.0, c=10.0, t=1.0, refrac=20),
    LIFNeuronLayer.C(r=1.0, c=20.0, t=1.0, refrac=20)
])

model = BPNNModel.compiled_model(model_input_shape, [nl], [])
```

# Run the model for 1000 timesteps

```python
traces = []

def end_time_step_callback(results):
    traces.append(results)
    
data = (np.ones(3,)*(1.5 if i > 200 else 0) for i in range(0, 1000))
    
model.run_time(data, end_time_step_callback)
```

# Extract the data we want and display

```python
neuron_layer_0_traces = np.array([r[0] for r in traces])

rend.render_figure([rend.LIFNeuronTraceRenderer(neuron_layer_0_traces, 'Layer 0 Neuron')], 0, 1000)
```
