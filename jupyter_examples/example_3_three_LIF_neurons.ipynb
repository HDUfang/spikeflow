# Three Leaky Integrate-and-fire neurons

```python
import numpy as np
from spikeflow import BPNNModel, LIFNeuronLayer
import spikeflow.drawing_utils.trace_renderers as rend
```

# Create a model layer with 3 neurons in a single layer

```python
import random
s = 47
random.seed(s)
np.random.seed(s)
```

```python
model_input_shape = (3,)

nl = LIFNeuronLayer.layer_from_tuples([
    LIFNeuronLayer.C(resistance=1.0, tau=10.0, threshhold=1.0, n_refrac=30),
    LIFNeuronLayer.C(resistance=1.0, tau=10.0, threshhold=1.0, n_refrac=20),
    LIFNeuronLayer.C(resistance=1.0, tau=20.0, threshhold=1.0, n_refrac=20)
])

model = BPNNModel.compiled_model(model_input_shape, [nl], [])
```

# Run the model for 1000 timesteps

```python
traces = []

def end_time_step_callback(i, graph, sess, results):
    traces.append(results)
    
data = (np.ones(3,)*(1.5 if (i > 200 and i < 1800) else 0) for i in range(0, 2000))
    
model.run_time(data, end_time_step_callback)
```

# Extract the data we want and display

```python
neuron_layer_0_traces = np.array([r[0] for r in traces])

rend.render_figure([rend.LIFNeuronTraceRenderer(neuron_layer_0_traces, 'Layer 0 Neuron')], 
                   0, 
                   2000)
```
